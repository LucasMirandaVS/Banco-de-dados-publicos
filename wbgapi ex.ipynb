```python
# Coletando os dados
# tabela escolhida : Poverty headcount ratio at $5.50 a day (2011 PPP) (% of population)
wb.data.DataFrame('SI.POV.UMIC', time=range(2010,2015), labels=True)
```

```python
# Agregando tabelas
wb.data.DataFrame(['SH.UHC.NOP2.ZS', 'SI.POV.DDAY'], 
                    economy = wb.region.members('LAC'), # América Latina
                    mrnev = 1, # Valores não vazios mais recentes ( 1 período)
                    timeColumns = True,
                    labels = True
                    )
```
